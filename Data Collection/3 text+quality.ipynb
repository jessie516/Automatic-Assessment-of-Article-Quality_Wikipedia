### Combining text and quality data

```python
text = pd.read_csv('/content/drive/MyDrive/학교/Dissertation/Data Collection/text.csv')
text.drop(['Unnamed: 0'], axis=1, inplace=True)
df_quality = pd.read_csv('/content/drive/MyDrive/학교/Dissertation/Data Collection/quality_api.csv')
df_quality.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
combined = text.merge(df_quality, on='title')
```

```python
combined.drop(combined[combined['page_id_text']!=combined['page_id_api']].index, inplace=True)
combined.reset_index(drop=True, inplace=True)
combined
```

### Preprocessing

__remove ones where 'quality_api' and 'quality_text' do not match__

__remove ones starting with '#redirect#', #Redirect', '{{wikitionary' or '{{wiktionary'__
