### Balancing article quality classes
- To balance article quality classes, a certain number of articles for each class is selected
- 1,300 articles for each of FA and GA-Class
- 650 articles for each of B, C, Start, and Stub-Class

```python
import pandas as pd
```

```python
reliable = pd.read_csv('/content/drive/MyDrive/학교/Dissertation/Data Collection2/reliable.csv')
reliable
```

```python
non_reliable = pd.read_csv('/content/drive/MyDrive/학교/Dissertation/Data Collection2/non_reliable.csv')
non_reliable
```

```python
reliable['quality'].value_counts()
```

```python
non_reliable['quality'].value_counts()
```

1. FA

```python
FA = reliable[reliable['quality']=='FA']
FA
```

```python
FA2 = FA.sample(n=1300, random_state=1)
FA2
```

2. GA

```python
GA = reliable[reliable['quality']=='GA']
GA
```

```python
GA2 = GA.sample(n=1300, random_state=1)
GA2
```

3. B-class

```python
B = non_reliable[non_reliable['quality']=='B']
B
```

```python
B2 = B.sample(n=650, random_state=1)
B2
```

4. C-class

```python
C = non_reliable[non_reliable['quality']=='C']
C
```

```python
C2 = C.sample(n=650, random_state=1)
C2
```

5. Start-class

```python
Start = non_reliable[non_reliable['quality']=='Start']
Start
```

```python
Start2 = Start.sample(n=650, random_state=1)
Start2
```

6. Stub-class

```python
Stub = non_reliable[non_reliable['quality']=='Stub']
Stub
```

```python
Stub2 = Stub.sample(n=650, random_state=1)
Stub2
```

----------

```python
data = pd.concat([FA2, GA2, B2, C2, Start2, Stub2])
data.reset_index(drop=True, inplace=True)
data
```

```python
data.to_csv('/content/drive/MyDrive/학교/Dissertation/Data Collection2/dataset.csv', index=False)
```
