```python
add_features = pd.read_csv('/content/drive/MyDrive/학교/Dissertation/Features/all_features2.csv')
add_features.drop(['Unnamed: 0'], axis=1, inplace=True)
add_features
```

__structure features__
- citation per text length
- link per text length
- image per text length
- abstract-text length ratio

__review features__
- review per day
- review per user

```python
add_features['citation_per_text'] = add_features['citation_count']/add_features['char_count']
add_features['link_per_text'] = (add_features['external_link_count']+add_features['internal_link_count'])/add_features['char_count']
add_features['image_per_text'] = add_features['image_count']/add_features['char_count']
add_features['abstract_text_ratio'] = add_features['abstract_size']/add_features['char_count']

add_features['review_per_day'] = add_features['num_edits']/add_features['article_age_days']
add_features['review_per_user'] = add_features['num_edits']/add_features['num_editors']
```
